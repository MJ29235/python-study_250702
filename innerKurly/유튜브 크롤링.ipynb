```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import csv, time, itertools

# ================== YouTube API 연결 ==================
API_KEY = "AIzaSyB4kcxtKFvp6Jx1TB3XmF_FFFUM_X2s-Hg"  # ← 유효한 API 키로 교체
youtube = build("youtube", "v3", developerKey=API_KEY)

# ================== 유튜브 영상 ID 목록 ==================
id_code = [
    '2HvZcFKMAhE',  # 댓글 1125개
    'GNMjHSHvikQ',  # 댓글 818개
    'oqkCD_tZe7s',  # 댓글 755개
    'npDo7RyGefU',  # 댓글 752개
    'P2ehMpPw72A',  # 댓글 559개
    'BH3pzRNyx_c',  # 댓글 387개
    'kwuw5WMmZwY',  # 댓글 283개
    'eOAXQXWowJo',  # 댓글 278개
    'V6-OaLPr0iM',  # 댓글 270개
    'NaSnTBmh2Kg'   # 댓글 225개
]

# ================== 유틸: 제목 일괄 조회(배치, 쿼터 절약) ==================
def get_titles_map(video_ids):
    """
    여러 video_id의 제목을 최대 50개 단위로 배치 조회.
    반환: {video_id: title}
    """
    titles = {}
    for chunk in (video_ids[i:i+50] for i in range(0, len(video_ids), 50)):
        try:
            req = youtube.videos().list(part="snippet", id=",".join(chunk))
            res = req.execute()
            for item in res.get("items", []):
                vid = item["id"]
                title = item["snippet"]["title"]
                titles[vid] = title
        except HttpError as e:
            print(f"[WARN] 제목 조회 실패(배치): {e}")
        time.sleep(0.2)  # API 완화
    return titles

# ================== 댓글 전체 크롤링(상위 댓글 전부) ==================
def crawl_all_comments(video_ids, order="relevance"):
    """
    각 video_id의 '상위 댓글(Top-level)'을 전부 페이지네이션하여 수집.
    (답글(replies)은 포함하지 않음. 필요 시 별도 함수로 확장 가능)
    반환: [(video_id, video_title, comment_text), ...]
    """
    data = []
    titles = get_titles_map(video_ids)

    for vid in video_ids:
        title = titles.get(vid, "(제목 불러오기 실패)")
        print(f"🎬 [{vid}] {title} → 모든 상위 댓글 수집 시작...")

        collected = 0
        page_token = None

        while True:
            try:
                req = youtube.commentThreads().list(
                    part="snippet",
                    videoId=vid,
                    maxResults=1000,          # API 한도 최대
                    textFormat="plainText",
                    order=order,
                    pageToken=page_token
                )
                res = req.execute()
            except HttpError as e:
                print(f"[WARN] {vid} 페이지 요청 실패: {e}")
                break

            items = res.get("items", [])
            for item in items:
                snippet = item.get("snippet", {})
                top = snippet.get("topLevelComment", {}).get("snippet", {})
                text = top.get("textDisplay", "")
                if text:
                    data.append((vid, title, text))
                    collected += 1

            page_token = res.get("nextPageToken")
            if not page_token:
                break

            # 과한 속도 방지
            time.sleep(0.2)

        print(f"✅ [{vid}] 상위 댓글 총 {collected}개 수집 완료")

        # 영상 사이 간단한 쿨다운
        time.sleep(0.3)

    return data

# ================== 실행 & CSV 저장 ==================
all_comments = crawl_all_comments(id_code, order="relevance")

csv_path = "youtube_all_top_level_comments.csv"
with open(csv_path, "w", newline="", encoding="utf-8-sig") as f:
    w = csv.writer(f)
    w.writerow(["video_id", "video_title", "comment_text"])
    w.writerows(all_comments)

print(f"\n💾 CSV 저장 완료 → {csv_path}")
print(f"총 수집된 댓글 수: {len(all_comments)}개")
```
