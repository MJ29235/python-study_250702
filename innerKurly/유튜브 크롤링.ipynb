```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import csv, time, itertools

# ================== YouTube API Ïó∞Í≤∞ ==================
API_KEY = "AIzaSyB4kcxtKFvp6Jx1TB3XmF_FFFUM_X2s-Hg"  # ‚Üê Ïú†Ìö®Ìïú API ÌÇ§Î°ú ÍµêÏ≤¥
youtube = build("youtube", "v3", developerKey=API_KEY)

# ================== Ïú†ÌäúÎ∏å ÏòÅÏÉÅ ID Î™©Î°ù ==================
id_code = [
    '2HvZcFKMAhE',  # ÎåìÍ∏Ä 1125Í∞ú
    'GNMjHSHvikQ',  # ÎåìÍ∏Ä 818Í∞ú
    'oqkCD_tZe7s',  # ÎåìÍ∏Ä 755Í∞ú
    'npDo7RyGefU',  # ÎåìÍ∏Ä 752Í∞ú
    'P2ehMpPw72A',  # ÎåìÍ∏Ä 559Í∞ú
    'BH3pzRNyx_c',  # ÎåìÍ∏Ä 387Í∞ú
    'kwuw5WMmZwY',  # ÎåìÍ∏Ä 283Í∞ú
    'eOAXQXWowJo',  # ÎåìÍ∏Ä 278Í∞ú
    'V6-OaLPr0iM',  # ÎåìÍ∏Ä 270Í∞ú
    'NaSnTBmh2Kg'   # ÎåìÍ∏Ä 225Í∞ú
]

# ================== Ïú†Ìã∏: Ï†úÎ™© ÏùºÍ¥Ñ Ï°∞Ìöå(Î∞∞Ïπò, ÏøºÌÑ∞ Ï†àÏïΩ) ==================
def get_titles_map(video_ids):
    """
    Ïó¨Îü¨ video_idÏùò Ï†úÎ™©ÏùÑ ÏµúÎåÄ 50Í∞ú Îã®ÏúÑÎ°ú Î∞∞Ïπò Ï°∞Ìöå.
    Î∞òÌôò: {video_id: title}
    """
    titles = {}
    for chunk in (video_ids[i:i+50] for i in range(0, len(video_ids), 50)):
        try:
            req = youtube.videos().list(part="snippet", id=",".join(chunk))
            res = req.execute()
            for item in res.get("items", []):
                vid = item["id"]
                title = item["snippet"]["title"]
                titles[vid] = title
        except HttpError as e:
            print(f"[WARN] Ï†úÎ™© Ï°∞Ìöå Ïã§Ìå®(Î∞∞Ïπò): {e}")
        time.sleep(0.2)  # API ÏôÑÌôî
    return titles

# ================== ÎåìÍ∏Ä Ï†ÑÏ≤¥ ÌÅ¨Î°§ÎßÅ(ÏÉÅÏúÑ ÎåìÍ∏Ä Ï†ÑÎ∂Ä) ==================
def crawl_all_comments(video_ids, order="relevance"):
    """
    Í∞Å video_idÏùò 'ÏÉÅÏúÑ ÎåìÍ∏Ä(Top-level)'ÏùÑ Ï†ÑÎ∂Ä ÌéòÏù¥ÏßÄÎÑ§Ïù¥ÏÖòÌïòÏó¨ ÏàòÏßë.
    (ÎãµÍ∏Ä(replies)ÏùÄ Ìè¨Ìï®ÌïòÏßÄ ÏïäÏùå. ÌïÑÏöî Ïãú Î≥ÑÎèÑ Ìï®ÏàòÎ°ú ÌôïÏû• Í∞ÄÎä•)
    Î∞òÌôò: [(video_id, video_title, comment_text), ...]
    """
    data = []
    titles = get_titles_map(video_ids)

    for vid in video_ids:
        title = titles.get(vid, "(Ï†úÎ™© Î∂àÎü¨Ïò§Í∏∞ Ïã§Ìå®)")
        print(f"üé¨ [{vid}] {title} ‚Üí Î™®Îì† ÏÉÅÏúÑ ÎåìÍ∏Ä ÏàòÏßë ÏãúÏûë...")

        collected = 0
        page_token = None

        while True:
            try:
                req = youtube.commentThreads().list(
                    part="snippet",
                    videoId=vid,
                    maxResults=1000,          # API ÌïúÎèÑ ÏµúÎåÄ
                    textFormat="plainText",
                    order=order,
                    pageToken=page_token
                )
                res = req.execute()
            except HttpError as e:
                print(f"[WARN] {vid} ÌéòÏù¥ÏßÄ ÏöîÏ≤≠ Ïã§Ìå®: {e}")
                break

            items = res.get("items", [])
            for item in items:
                snippet = item.get("snippet", {})
                top = snippet.get("topLevelComment", {}).get("snippet", {})
                text = top.get("textDisplay", "")
                if text:
                    data.append((vid, title, text))
                    collected += 1

            page_token = res.get("nextPageToken")
            if not page_token:
                break

            # Í≥ºÌïú ÏÜçÎèÑ Î∞©ÏßÄ
            time.sleep(0.2)

        print(f"‚úÖ [{vid}] ÏÉÅÏúÑ ÎåìÍ∏Ä Ï¥ù {collected}Í∞ú ÏàòÏßë ÏôÑÎ£å")

        # ÏòÅÏÉÅ ÏÇ¨Ïù¥ Í∞ÑÎã®Ìïú Ïø®Îã§Ïö¥
        time.sleep(0.3)

    return data

# ================== Ïã§Ìñâ & CSV Ï†ÄÏû• ==================
all_comments = crawl_all_comments(id_code, order="relevance")

csv_path = "youtube_all_top_level_comments.csv"
with open(csv_path, "w", newline="", encoding="utf-8-sig") as f:
    w = csv.writer(f)
    w.writerow(["video_id", "video_title", "comment_text"])
    w.writerows(all_comments)

print(f"\nüíæ CSV Ï†ÄÏû• ÏôÑÎ£å ‚Üí {csv_path}")
print(f"Ï¥ù ÏàòÏßëÎêú ÎåìÍ∏Ä Ïàò: {len(all_comments)}Í∞ú")
```
