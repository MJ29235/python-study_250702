```python
# so, selenium으로 크롤링하는 방법!!
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
from openpyxl import Workbook


#1. selenium 기본 옵션 설정
options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--headless")

driver = webdriver.Chrome(options = options)

#2. 페이지 접속
url = "https://www.kurly.com/goods/5090278"
driver.get(url)

time.sleep(5) #페이지 로딩 대기시간

#3.HTML 가져오기
soup = BeautifulSoup(driver.page_source, "html.parser")

# 4. 리뷰 텍스트 추출하기
reviews = soup.select("p.css-y49dcn.e36z05c13")
# 리뷰즈의 값을 찾아와서 값을 하나씩 저장하기
review_lists =[review.get_text().strip() for review in reviews] 

driver.quit() #종료를 시키지 않으면 계속해서 행동을 하기때문에 봇은 이상함을 감지함.

wb = Workbook()
ws = wb.active
ws.title = "Kurly Reviews"

ws.append(["리뷰 내용"])
for item in review_lists:
    ws.append([item])

wb.save("kurly_reviews.xlsx")
print("엑셀저장완료 : kurly_reviews.xlsx")
```
