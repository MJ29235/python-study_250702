```python
#ì²«ë²ˆì§¸ í˜ì´ì§€ ë°ì´í„°ë§Œ ê°€ì ¸ì˜´
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pickle
import time

# ì˜µì…˜ ì„¤ì •
options = Options()
prefs = {
    "profile.default_content_setting_values.notifications": 1  # 1:í—ˆìš©, 2:ì°¨ë‹¨
}
options.add_experimental_option("prefs", prefs)
options.add_argument("--start-maximized")
options.add_argument("window-size=1920x1080")
options.add_argument("lang=ko_KR")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 1. í˜ì´ìŠ¤ë¶ ë©”ì¸ í˜ì´ì§€ ì—´ê¸° (ì¿ í‚¤ë¥¼ ë„£ê¸° ìœ„í•´ ê¸°ë³¸ í˜ì´ì§€ë¥¼ ë¨¼ì € ì—´ì–´ì•¼ í•¨)
driver.get("https://www.facebook.com")

# 2. ì¿ í‚¤ ë¶ˆëŸ¬ì˜¤ê¸°
with open("facebook_cookies.pkl", "rb") as f:
    cookies = pickle.load(f)

for cookie in cookies:
    # Facebook ì¿ í‚¤ ì¤‘ domainì´ .facebook.comì´ ì•„ë‹ˆê±°ë‚˜ ìœ íš¨í•˜ì§€ ì•Šì€ ì¿ í‚¤ëŠ” ì œì™¸í•  ìˆ˜ë„ ìˆìŒ
    if 'sameSite' in cookie:
        if cookie['sameSite'] == 'None':
            cookie['sameSite'] = 'Strict'  # Selenium í˜¸í™˜ ìœ„í•´ ë³€ê²½
    driver.add_cookie(cookie)

# 3. í˜ì´ì§€ ìƒˆë¡œê³ ì¹¨í•´ì„œ ì¿ í‚¤ ì ìš© (ë¡œê·¸ì¸ ìƒíƒœ ìœ ì§€ë¨)
driver.refresh()

time.sleep(5)  # í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°

# 4. íŠ¹ì • í˜ì´ì§€ URLë¡œ ì´ë™ (ì˜ˆ: ìœ ëª…ì¸ í˜ì´ì§€ë‚˜ ì»¤ë®¤ë‹ˆí‹° í˜ì´ì§€)
target_page_url = "https://www.facebook.com/koreaswim"
driver.get(target_page_url)

time.sleep(5)  # í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°

# 5. ê²Œì‹œê¸€ ì œëª© í¬ë¡¤ë§
# í˜ì´ìŠ¤ë¶ ê²Œì‹œê¸€ì€ êµ¬ì¡°ê°€ ìì£¼ ë°”ë€Œë¯€ë¡œ, í˜„ì¬ (2025ë…„ ê¸°ì¤€) ê²Œì‹œê¸€ ì œëª©ì— ì ‘ê·¼í•˜ëŠ” ì˜ˆì‹œ CSS ì„ íƒìì…ë‹ˆë‹¤.
# ì¼ë°˜ì ìœ¼ë¡œ ê²Œì‹œê¸€ í…ìŠ¤íŠ¸ëŠ” div[data-pagelet^="FeedUnit"] ë‚´ë¶€ì— ìˆìŠµë‹ˆë‹¤.
# í•„ìš”ì— ë”°ë¼ ì ì ˆíˆ ì„ íƒìë¥¼ ë³€ê²½í•´ì•¼ í•©ë‹ˆë‹¤.

posts = driver.find_elements(By.CSS_SELECTOR, "div[data-pagelet^='TimelineFeedUnit_']")
# print(posts)

for i, post in enumerate(posts, 1) :
    print(f"{i}. {post.text}")

driver.quit()
```

```python
#ìŠ¤í¬ë¡¤ í•´ì„œ ê°€ì ¸ì˜¤ì§€ë§Œ ì„ íƒì ë™ì ìœ¼ë¡œ ë°”ê»´ì„œ ë°ì´í„° ëª»ê°€ì ¸ì˜¤ëŠ” ì¤‘ ìˆ˜ì •í•´ì•¼í•¨
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pickle
import time

# ì˜µì…˜ ì„¤ì •
options = Options()
prefs = {
    "profile.default_content_setting_values.notifications": 1  # ì•Œë¦¼ í—ˆìš©
}
options.add_experimental_option("prefs", prefs)
options.add_argument("--start-maximized")
options.add_argument("window-size=1920x1080")
options.add_argument("lang=ko_KR")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 1. í˜ì´ìŠ¤ë¶ ë©”ì¸ í˜ì´ì§€ ì—´ê¸° (ì¿ í‚¤ë¥¼ ë„£ê¸° ìœ„í•´)
driver.get("https://www.facebook.com")

# 2. ì¿ í‚¤ ë¶ˆëŸ¬ì˜¤ê¸°
with open("facebook_cookies.pkl", "rb") as f:
    cookies = pickle.load(f)

for cookie in cookies:
    if 'sameSite' in cookie:
        if cookie['sameSite'] == 'None':
            cookie['sameSite'] = 'Strict'
    driver.add_cookie(cookie)

# 3. í˜ì´ì§€ ìƒˆë¡œê³ ì¹¨í•´ì„œ ì¿ í‚¤ ì ìš©
driver.refresh()
time.sleep(5)

# 4. íŠ¹ì • í˜ì´ì§€ ì´ë™
target_page_url = "https://www.facebook.com/koreaswim"
driver.get(target_page_url)
time.sleep(5)

# 5. ìŠ¤í¬ë¡¤ ë°˜ë³µ ì„¤ì •
SCROLL_PAUSE_TIME = 3  # ì´ˆ ë‹¨ìœ„ ëŒ€ê¸° ì‹œê°„
MAX_SCROLLS = 10       # ìŠ¤í¬ë¡¤ ë°˜ë³µ íšŸìˆ˜

last_height = driver.execute_script("return document.body.scrollHeight")

for i in range(MAX_SCROLLS):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        print("ë” ì´ìƒ ìŠ¤í¬ë¡¤ í•  ë‚´ìš©ì´ ì—†ìŠµë‹ˆë‹¤.")
        break
    last_height = new_height

# 6. ê²Œì‹œê¸€ ìˆ˜ì§‘ ë° ìµœëŒ€ 50ê°œê¹Œì§€ë§Œ ì¶œë ¥
max_posts = 50
posts = driver.find_elements(By.CSS_SELECTOR, "div[data-pagelet^='TimelineFeedUnit_']")

print(f"ì „ì²´ ê²Œì‹œê¸€ ìˆ˜: {len(posts)}")

for idx, post in enumerate(posts[:max_posts], 1):
    print(f"{idx}. {post.text[:100]}")  # ê²Œì‹œê¸€ ì• 100ì ì¶œë ¥

driver.quit()
```
