```python
# 페이스북 로그인 후 특정 페이지의 최근 게시글 제목을 수집하세요.

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pickle#쿠키값이나 여러 정보값 가져올 수 있게 하는 라이브러리
import time

options = Options()
options.add_argument("--start-maximized")
options.add_argument("window-size=1920x1080")
options.add_argument("lang=ko_KR")

service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service, options=options)

url = "https://www.facebook.com/login"
driver.get(url)

print("수동으로 인증 시작!")
time.sleep(30)

#cookie 값을 저장하고 저장한 이름으로 오픈해서 실제 값을 넣어줌
with open("facebook_cookies.pkl", "wb") as f :#앞에 선언된 이름으로 실행해라
    pickle.dump(driver.get_cookies(), f)#쿠키 값 받아오기와서 윗줄로 보냄 dump => 받아온 쿠키값 정렬

print("쿠키 저장 완료")
driver.quit()
```

```python
#첫번째 페이지 데이터만 가져옴
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pickle
import time

# 옵션 설정
options = Options()
prefs = {
    "profile.default_content_setting_values.notifications": 1  # 1:허용, 2:차단
}
options.add_experimental_option("prefs", prefs)
options.add_argument("--start-maximized")
options.add_argument("window-size=1920x1080")
options.add_argument("lang=ko_KR")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 1. 페이스북 메인 페이지 열기 (쿠키를 넣기 위해 기본 페이지를 먼저 열어야 함)
driver.get("https://www.facebook.com")

# 2. 쿠키 불러오기
with open("facebook_cookies.pkl", "rb") as f:
    cookies = pickle.load(f)

for cookie in cookies:
    # Facebook 쿠키 중 domain이 .facebook.com이 아니거나 유효하지 않은 쿠키는 제외할 수도 있음
    if 'sameSite' in cookie:
        if cookie['sameSite'] == 'None':
            cookie['sameSite'] = 'Strict'  # Selenium 호환 위해 변경
    driver.add_cookie(cookie)

# 3. 페이지 새로고침해서 쿠키 적용 (로그인 상태 유지됨)
driver.refresh()

time.sleep(5)  # 페이지 로딩 대기

# 4. 특정 페이지 URL로 이동 (예: 유명인 페이지나 커뮤니티 페이지)
target_page_url = "https://www.facebook.com/koreaswim"
driver.get(target_page_url)

time.sleep(5)  # 페이지 로딩 대기

# 5. 게시글 제목 크롤링
# 페이스북 게시글은 구조가 자주 바뀌므로, 현재 (2025년 기준) 게시글 제목에 접근하는 예시 CSS 선택자입니다.
# 일반적으로 게시글 텍스트는 div[data-pagelet^="FeedUnit"] 내부에 있습니다.
# 필요에 따라 적절히 선택자를 변경해야 합니다.

posts = driver.find_elements(By.CSS_SELECTOR, "div[data-pagelet^='TimelineFeedUnit_']")
# print(posts)

for i, post in enumerate(posts, 1) :
    print(f"{i}. {post.text}")

driver.quit()
```

```python
#스크롤 해서 가져오지만 선택자 동적으로 바껴서 데이터 못가져오는 중 수정해야함
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pickle
import time

# 옵션 설정
options = Options()
prefs = {
    "profile.default_content_setting_values.notifications": 1  # 알림 허용
}
options.add_experimental_option("prefs", prefs)
options.add_argument("--start-maximized")
options.add_argument("window-size=1920x1080")
options.add_argument("lang=ko_KR")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 1. 페이스북 메인 페이지 열기 (쿠키를 넣기 위해)
driver.get("https://www.facebook.com")

# 2. 쿠키 불러오기
with open("facebook_cookies.pkl", "rb") as f:
    cookies = pickle.load(f)

for cookie in cookies:
    if 'sameSite' in cookie:
        if cookie['sameSite'] == 'None':
            cookie['sameSite'] = 'Strict'
    driver.add_cookie(cookie)

# 3. 페이지 새로고침해서 쿠키 적용
driver.refresh()
time.sleep(5)

# 4. 특정 페이지 이동
target_page_url = "https://www.facebook.com/koreaswim"
driver.get(target_page_url)
time.sleep(5)

# 5. 스크롤 반복 설정
SCROLL_PAUSE_TIME = 3  # 초 단위 대기 시간
MAX_SCROLLS = 10       # 스크롤 반복 횟수

last_height = driver.execute_script("return document.body.scrollHeight")

for i in range(MAX_SCROLLS):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        print("더 이상 스크롤 할 내용이 없습니다.")
        break
    last_height = new_height

# 6. 게시글 수집 및 최대 50개까지만 출력
max_posts = 50
posts = driver.find_elements(By.CSS_SELECTOR, "div[data-pagelet^='TimelineFeedUnit_']")

print(f"전체 게시글 수: {len(posts)}")

for idx, post in enumerate(posts[:max_posts], 1):
    print(f"{idx}. {post.text[:100]}")  # 게시글 앞 100자 출력

driver.quit()
```
