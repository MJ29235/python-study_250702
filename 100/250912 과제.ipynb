```python
import requests
from bs4 import BeautifulSoup
import re  # 숫자 추출을 위한 정규표현식

url = "https://books.toscrape.com/catalogue/category/books/travel_2/index.html"
base_url = "https://books.toscrape.com/catalogue/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response.encoding = 'utf-8'  # 인코딩 명시

soup = BeautifulSoup(response.text, "html.parser")
books = soup.select("article.product_pod")
category = "Travel"

for book in books:
    title = book.h3.a["title"]
    
    price_text = book.select_one("p.price_color").text.strip()
    price_number = re.findall(r"\d+\.\d+", price_text)  # 숫자만 추출
    price = float(price_number[0]) if price_number else 0.0

    stock = book.select_one("p.instock.availability").text.strip()
    rating_tag = book.select_one("p.star-rating")
    rating = rating_tag["class"][1] if rating_tag else "No rating"
    relative_url = book.h3.a["href"]
    product_url = base_url + relative_url.replace("../../../", "")
    
    cursor.execute("""
        INSERT INTO product (CODE, TITLE, PRICE, STOCK, RATING, CATEGORY, URL)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
    """, (title, price, stock, rating, category, product_url))

db.commit()
db.close()
```

```python
import requests
from bs4 import BeautifulSoup
import pymysql
import re

# 1. DB 연결
db = pymysql.connect(
    host="localhost",
    port=3306,
    user="root",
    passwd="asdf1122",
    charset="utf8"
)
cursor = db.cursor()

# 2. DB 선택 및 테이블 생성
cursor.execute("CREATE DATABASE IF NOT EXISTS ecommerce CHARACTER SET utf8mb4 COLLATE utf8mb4_unicode_ci;")
cursor.execute("USE ecommerce;")

cursor.execute("DROP TABLE IF EXISTS product;")
cursor.execute("""
    CREATE TABLE product (
        code VARCHAR(225) NOT NULL PRIMARY KEY,
        title VARCHAR(200) NOT NULL,
        price INT,
        stock VARCHAR(50),
        rating VARCHAR(50),
        category VARCHAR(50),
        url VARCHAR(255)
    );
""")
db.commit()

# 3. 웹 크롤링
url = "https://books.toscrape.com/catalogue/category/books/travel_2/index.html"
base_url = "https://books.toscrape.com/catalogue/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response.encoding = "utf-8"
soup = BeautifulSoup(response.text, "html.parser")

books = soup.select("article.product_pod")
category = "Travel"

for book in books:
    title = book.h3.a["title"]

    price_text = book.select_one("p.price_color").text.strip()
    price_number = re.findall(r"\d+\.\d+", price_text)
    price = int(float(price_number[0])) if price_number else 0

    stock = book.select_one("p.instock.availability").text.strip()
    rating_tag = book.select_one("p.star-rating")
    rating = rating_tag["class"][1] if rating_tag else "No rating"
    relative_url = book.h3.a["href"]
    product_url = base_url + relative_url.replace("../../../", "")

    # 🔑 고유 코드 (_id 역할)
    code = product_url.split("/")[-2]

    # 7개 값 삽입
    cursor.execute("""
        INSERT INTO product (code, title, price, stock, rating, category, url)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
    """, (
        code, title, price, stock, rating, category, product_url
    ))

db.commit()
db.close()
```
