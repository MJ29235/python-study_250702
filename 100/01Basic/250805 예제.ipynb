```python
#구글 Scholar에서 'SEO Marketing' 논문 제목을 5개 수집하세요

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
service = Service(ChromeDriverManager().install())
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--lang=ko_KR")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

driver = webdriver.Chrome(service=service,options=options)
query="?hl=ko&as_sdt=0%2C5&q=SEO+Marketing&btnG="
url=f"https://scholar.google.com/scholar{query}"
driver.get(url)

titles = driver.find_elements(By.CSS_SELECTOR,"h3.gs_rt > a")
for i,title in enumerate(titles[:5],1):
    print(f"{i}. {title.text.strip()}")

time.sleep(2)
driver.quit()
```

```python
# TIOBE Index 사이트에서 상위 10개 프로그래밍 언어 이름을 수집하세요.

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
service = Service(ChromeDriverManager().install())
options.add_argument('--headless')
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--lang=ko_KR")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

driver = webdriver.Chrome(service=service,options=options)
query="tiobe-index/"
url=f"https://www.tiobe.com/{query}"
driver.get(url)

driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
titles = driver.find_elements(By.CSS_SELECTOR,"td.td-top20 > td")
for title in titles:
    print(f"{title.text.strip().replace("\n","")}")

time.sleep(1)
driver.quit()
```

```python
# 데이터마켓(data.go.kr)에서 '소비' 관련 공공데이터 제목을 10개 수집하세요

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
service = Service(ChromeDriverManager().install())
options.add_argument('--headless')
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--lang=ko_KR")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

driver = webdriver.Chrome(service=service,options=options)
query="tcs/dss/selectDataSetList.do?keyword=소비&brm=&svcType=&recmSe=N&conditionType=init&extsn=&kwrdArray="
url=f"https://www.data.go.kr/{query}"
driver.get(url)

driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
titles = driver.find_elements(By.CSS_SELECTOR,"span.title")
for i, title in enumerate(titles[:10],1):
    print(f"{i}. {title.text.strip().replace("\n","")}")

time.sleep(1)
driver.quit()
```

```python
# 알라딘에서 '자기계발' 베스트 도서의 제목과 저자를 수집하세요

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
service = Service(ChromeDriverManager().install())
options.add_argument('--headless')
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--lang=ko_KR")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

driver = webdriver.Chrome(service=service,options=options)
query="/shop/wbrowse.aspx?CID=336&BrowseTarget=BestSeller"
url=f"https://www.aladin.co.kr/{query}"
driver.get(url)

driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
texts = driver.find_elements(By.CSS_SELECTOR,"td.tc")
for i, text in enumerate(texts,1):
    title = text.find_element(By.CSS_SELECTOR,"a.bo")
    author = text.find_element(By.CSS_SELECTOR,"span.gw")
    print(f"{i}위. {title.text.strip().replace("\n","")} / {author.text.strip().replace("\n","")}")

time.sleep(1)
driver.quit()
```
