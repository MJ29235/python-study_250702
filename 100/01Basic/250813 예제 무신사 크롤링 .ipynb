```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import re
import pymysql

# 드라이버 세팅
service = Service(ChromeDriverManager().install())
options = Options()
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--window-size=1920x1080')
options.add_argument('--start-maximized')
options.add_argument('--lang=ko_KR')

driver = webdriver.Chrome(service=service, options=options)

# 무신사 랭킹 페이지 접속
driver.get("https://www.musinsa.com/main/musinsa/ranking?gf=M&storeCode=musinsa&sectionId=200&contentsId=&categoryCode=000&ageBand=AGE_BAND_ALL")
time.sleep(3)

# 스크롤 내려서 50개 이상 로딩
scroll_pause = 1
for _ in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause)

# 상위 50개 상품
products = driver.find_elements(By.CSS_SELECTOR, "div.sc-1t5ihy5-1.jzppnI")[:50]

product_data = []

for p in products:
    try:
        name = p.find_element(By.CSS_SELECTOR, "span.text-title_18px_med.sc-1omefes-1.exqQRL.font-pretendard").text.strip()               # 상품명
        brand = p.find_element(By.CSS_SELECTOR, "span.text-body_14px_med.font-pretendard").text.strip()            # 브랜드명
        price_text = p.find_element(By.CSS_SELECTOR, "span.text-title_18px_semi.sc-1hw5bl8-7.kXhdZT.text-black.font-pretendard").text.strip()              # 가격 정보
        review_text = p.find_element(By.CSS_SELECTOR, "span.text-title_18px_semi.sc-1hw5bl8-6.hROMjI.text-red.font-pretendard").text.strip()          # 리뷰 수
        discount_text = p.find_element(By.CSS_SELECTOR, "span.text-body_13px_reg.underline.text-gray-600.font-pretendard").text.strip()

        # 숫자만 추출
        price = int(re.sub(r"[^\d]", "", price_text.split()[0])) if price_text else 0
        discount_rate = int(re.sub(r"[^\d]", "", discount_text)) if discount_text else 0
        review_count = int(re.sub(r"[^\d]", "", review_text)) if review_text else 0

        product_data.append({
            "상품명": name,
            "브랜드명": brand,
            "가격": price,
            "할인률": discount_rate,
            "리뷰수": review_count
        })

    except Exception as e:
        print("에러:", e)
        continue

driver.quit()

# DataFrame
df = pd.DataFrame(product_data)
print(df.head())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# 크롬 옵션 설정
options = Options()
options.add_argument("--headless")  # 창 안띄움
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--no-sandbox")

# 드라이버 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.musinsa.com/main/musinsa/ranking?gf=M&storeCode=musinsa")

time.sleep(3)

# 스크롤 다운 (상품 더 불러오기)
scroll_pause = 2
for _ in range (3):
    #스크롤 밑으로 내림
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause)
    
# 상품명 + URL 추출
product_elements = driver.find_elements(By.CSS_SELECTOR, "a.gtm-select-item")

data = []
for el in product_elements[:50]:
    name = el.text.strip()
    url = el.get_attribute("href")
    if name and url:
        data.append({"상품명": name, "URL": url})

# DataFrame 변환
df = pd.DataFrame(data)
print(df.head())

driver.quit()
```

```python
driver = webdriver.Chrome(service=service, options=options)
all_products = []

for idx, row in df.iterrows(): #iterrows : 1개의 행을 가져오고 반복할때마다 1개의 값을 찾아오게 하는 함수
    print(f"{idx + 1}번째 상품리뷰 수집 중 :", {row['상품명']})
    driver.get(row["URL"]) #row라는 행 안에 URL이라는 값에 매칭되는 애를 찾아오겠다
    time.sleep(2)

cards = driver.find_elements(By.CSS_SELECTOR,"div.sc-1t5ihy5-0.wvTSV.gtm-view-item-list")
for card in cards:
    try :
        productName = driver.find_elements(By.CSS_SELECTOR,"span.text-title_18px_med.sc-1omefes-1.exqQRL.font-pretendard")
        brandName = driver.find_elements(By.CSS_SELECTOR,"span.text-body_14px_med.font-pretendard")
        price = driver.find_elements(By.CSS_SELECTOR,"span.text-title_18px_semi.sc-1hw5bl8-7.kXhdZT.text-black.font-pretendard")
        disPercent = driver.find_elements(By.CSS_SELECTOR,"span.text-title_18px_semi.sc-1hw5bl8-6.hROMjI.text-red.font-pretendard")
        reviewNum = driver.find_elements(By.CSS_SELECTOR,"span.text-body_13px_reg.underline.text-gray-600.font-pretendard")
        for card in cards[:50] :
            text = card.text.strip()
            if text : 
                all_products.append({
                    "상품명" : row["상품명"], #row상품명에서 가져옴
                    "리뷰" : text
                }) # 비어있는 리스트에 값추가

    except Exception as e:
        print("리뷰 수집 실패 : ", e)
        continue

driver.quit()
from openpyxl import Workbook
wb = Workbook()
ws = wb.active
ws.title = "무신사 리뷰"

ws. append(["상품명","브랜드명","가격","할인률","리뷰 수"])

#각각의 리뷰안에 상품명 한쌍으로 만들기
for r in all_products:
    ws.append([r["상품명"],r["브랜드명"],r["가격"],r["할인률"],r["리뷰 수"]])

wb.save("musinsa_top50_ranking.xlsx")
print("⭐엑셀 저장완료 : musinsa_top50_ranking.xlsx")
```
