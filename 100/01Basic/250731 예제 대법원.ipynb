```python
# 위키백과에서 '마케팅' 문서의 첫 문단 요약을 추출하세요.

import requests
from bs4 import BeautifulSoup

query = "마케팅"
url = f"https://ko.wikipedia.org/wiki/{query}"

headers ={
    "User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text,"html.parser")

text = soup.select_one("div.mw-parser-output > P")
print(text.text)
```

```python
# 서울 열린데이터광장 공공데이터 목록에서 데이터셋 제목을 10개 수집하세요.


import requests
from bs4 import BeautifulSoup

query = "tcs/dss/selectDataSetList.do"
url = f"https://www.data.go.kr/{query}"

headers ={
    "User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text,"html.parser")

titles = soup.select("span.recent-title")
for i,title in enumerate(titles[:10],1):
    print(f"{i}. {title.get_text().strip().replace("\n","").replace("Update","")}")
```

```python
# 인터파크 티켓에서 인기 콘서트 제목과 일정을 수집하세요.

import requests
from bs4 import BeautifulSoup

query = "contents/genre/concert"
url = f"https://tickets.interpark.com/{query}"

headers ={
    "User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text,"html.parser")

titles = soup.select("li.TicketItem_goodsName__Ju76j")
for i,title in enumerate(titles,1):
    print(f"{i}. {title.get_text().strip().replace("\n","").replace("Update","")}")
```

```python
# 대법원 판례 검색 페이지에서 최근 등록된 판례 제목을 10개 수집하세요.

import requests
from bs4 import BeautifulSoup

query = "pgp/index.on?m=PGP1011M01&l=N&c=900"
url = f"https://portal.scourt.go.kr/{query}"

headers ={
    "User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text,"html.parser")

titles = soup.select("div.w2anchor.infoTit")
for i,title in enumerate(titles[:10],1):
    print(f"{i}. {title.get_text().strip().replace('\n','').replace('Update','')}")
```

```python
# 대법원 판례 검색 페이지에서 최근 등록된 판례 제목을 10개 수집하세요.

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())

options = Options()
options.add_argument("--headless")
options.add_argument("--start-maximized")
options.add_argument("--window-size=1920x1080")
options.add_argument("--lang=ko_KR")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

driver = webdriver.Chrome(service=service, options=options)
query = "pgp/index.on?m=PGP1011M01&l=N&c=900"
url = f"https://portal.scourt.go.kr/{query}"
driver.get(url)


elements = driver.find_elements(By.CSS_SELECTOR,"div.w2anchor.infoTit")
for i, element in enumerate(elements[:10], 1):
         print(f"{i}. {element.text.strip()}")


time.sleep(10)

driver.quit()
```

```python
# 삼성전자 뉴스룸에서 최근 기사 제목과 날짜를 수집하세요.

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())

options = Options()
options.add_argument("--headless")
options.add_argument("--start-maximized")
options.add_argument("--window-size=1920x1080")
options.add_argument("--lang=ko_KR")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

driver = webdriver.Chrome(service=service, options=options)
query = "category/기업"
url = f"https://news.samsung.com/kr/{query}"
driver.get(url)


datas = driver.find_elements(By.CSS_SELECTOR,"div.desc") 
titles = driver.find_elements(By.CSS_SELECTOR,"div.desc > span.title.ellipsis") 
dates = driver.find_elements(By.CSS_SELECTOR,"div.desc > span.date") 
for i, data in enumerate(datas,1):
    title = titles[i-1].text
    date = dates[i-1].text
    print(f"{i}. {title} / {date}")


time.sleep(1)

driver.quit()
```
