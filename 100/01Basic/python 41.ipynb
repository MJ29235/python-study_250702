### yes24 경제 베스트셀러 1~10 위 도서 제목과 저자를 수집해주세요

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.yes24.com/product/category/bestseller?categoryNumber=001001025"

headers = {
    "User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36'
}

response = requests.get(url, headers = headers)
response.encoding = "utf-8"
soup = BeautifulSoup(response.text,"html.parser")

books = soup.select("a.gd_name")
authors = soup.select("span.authPub.info_auth")
for i, book in enumerate(books[:10],1):
    book = book.text.strip()
    for i, author in enumerate(authors[:10],1):

        author = author.text.strip()
        print(f"{i}. {book} - {author}")
```

```python
#선생님해설
#잡코리아는 동적사이트!
# openpyxl -> 행열 주기 어려움
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
import urllib.parse # 1, 선 인코딩 시켜주는 라이브러리, 공백이나 특수문자를 이해할 수 있도록!
import pandas as pd # 2, pandas -> 수집된 정보를 전처리(이쁘게 다듬기)해서 시각화시켜주는 라이브러리! 테이블로 시각화 하는데 강점! 외부라이브러리. 사전설치 필수!

service = Service(ChromeDriverManager().install())
options = Options()
options.add_argument("--headless")
options.add_argument("--window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")

driver = webdriver.Chrome(service=service, options=options)
query = "데이터 분석가"
# 3, 사전적으로 빈칸을 인코딩 해주기
encoded_query = urllib.parse.quote(query) # 4, ()안에 인코딩하고자 하는 텍스트

# 5, 6개 페이지 가져오는 법
max_pages = 6
results = []
for page in range(1,max_pages+1) :
    url = f"https://www.jobkorea.co.kr/Search/?stext={encoded_query}&Page_No={page}&local=I000_&tabType=recruit&careerType=1"
    driver.get(url)
    time.sleep(2)
    # 6, 앵커 윗단계의 것을 기준으로 값을 가져옴!
    job_posts = driver.find_elements(By.CSS_SELECTOR,"div.styles_mb_space8__dk46ts1p")
    for post in job_posts :
        try :
            title_elem = post.find_element(By.CSS_SELECTOR,"a.sn28bt0")
            title = title_elem.text.strip()
            link = title_elem.get_attribute("href") # 7, 상위 단계의 것(앵커태그)이기 때문에 택스트와 url 동시에 찾아와서 분리 가능!
            results.append({
                "제목" : title,
                "제목 링크" : link
            })
        except :
            continue

driver.quit()
print(results)
# 8, 판다스로 엑셀 시각화 해보기!
df = pd.DataFrame(results) # 9, 열과 행의 반복된 패턴을 캐치해서 데이터 프레임 형식으로 알아서 정리해줌!
print(df)
# 10, 엑셀로 바로 출력도 가능
df.to_excel("잡코리아_데이터분석가_서울_신입_1페이지..xlsx", index = False)
print("excel 저장완료!")
```
