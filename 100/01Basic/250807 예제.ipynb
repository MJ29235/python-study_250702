```python
# 마켓컬리에서 신상품 중 평점이 높은 10개 상품의 이름과 평점을 수집하세요.

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
service = Service(ChromeDriverManager().install())
# options.add_argument("--headless")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--lang=ko_KR")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

driver = webdriver.Chrome(service=service,options=options)
query ="/collection-groups/market-newproduct?page=1&collection=newranking1"
url = f"https://www.kurly.com/{query}"

driver.get(url)

search_button = driver.find_element(By.CSS_SELECTOR,"span.css-1xyu7j9.e17w4cfr2")
search_button.click()
time.sleep(2)

titles = driver.find_elements(By.CSS_SELECTOR,"span.css-1dry2r1.e1c07x485")
for i,title in enumerate(titles[:10],1):
    print(f"{i}. {title.text.strip()}")
```

```python
# G마켓에서 '게이밍 키보드' 검색 후 스크롤을 내려 상품명과 리뷰수를 수집하세요.

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
service = Service(ChromeDriverManager().install())
options.add_argument("--headless")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--lang=ko_KR")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

driver = webdriver.Chrome(service=service,options=options)
query ="/n/search?keyword=게이밍키보드"
url = f"https://www.gmarket.co.kr{query}"

driver.get(url)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)

boxes = driver.find_elements(By.CSS_SELECTOR,"div.box__information")

for i, box in enumerate(boxes, 1):
    title = box.find_element(By.CSS_SELECTOR, "span.text__item").text.strip()
    review = box.find_element(By.CSS_SELECTOR, "span.text").text.strip()
    print(f"{i}. {title}'\n'리뷰 수: {review.replace("원","")}")
```
