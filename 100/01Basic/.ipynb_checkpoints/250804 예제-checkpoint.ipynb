```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# í¬ë¡¬ ì˜µì…˜ ì„¤ì •
options = Options()
options.add_argument("--headless")
options.add_argument("--window-size=1920,1080")
options.add_argument("--lang=ko_KR")
options.add_argument("--disable-blink-features=AutomationControlled")

# í¬ë¡¬ ë“œë¼ì´ë²„ ì„¤ì •
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 'ì¶”ì²œ ì—¬í–‰ì§€' ì§€ì—­ë³„ íŽ˜ì´ì§€ ì ‘ì†
base_url = "https://korean.visitkorea.or.kr/travel-theme/theme-list.html"
driver.get(base_url)
time.sleep(2)

# ì§€ì—­ë³„ íƒ­ í´ë¦­ (ì˜ˆ: ê°•ì›ë„, ê²½ê¸°ë„ ë“±)
region_buttons = driver.find_elements(By.CSS_SELECTOR, "ul.tab_theme li a")

for region_button in region_buttons:
    region_name = region_button.text.strip()
    print(f"\nâ–¶ {region_name}")

    region_button.click()
    time.sleep(2)

    # í•´ë‹¹ ì§€ì—­ì˜ ì—¬í–‰ì§€ ëª©ë¡ ìˆ˜ì§‘
    items = driver.find_elements(By.CSS_SELECTOR, "ul.list_thumType li")

    if not items:
        print("  (ë°ì´í„° ì—†ìŒ)")
        continue

    for i, item in enumerate(items, 1):
        title = item.find_element(By.CSS_SELECTOR, "div.info_area > strong.tit").text.strip()
        print(f"  {i}. {title}")

    time.sleep(1)

driver.quit()
```
