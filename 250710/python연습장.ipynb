```python
import requests
from bs4 import BeautifulSoup
url ="https://v.daum.net/v/20170615203441266"
res = requests.get(url)
soup = BeautifulSoup(res.content,"html.parser")

data = soup.select("body")
for datas in data:
    print(datas.get_text().strip().replace("\n",""))
```

```python
#복수의 페이지에서 값을 찾아오는 방법!!
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/" #파라미터가 빠졌는지 확인 필수!!
for page_num in range(7) :
    if page_num ==0 :
        res = requests.get(url) #0은 메인페이지로 간주
    else :
        res = requests.get(f"{url}page{page_num+1}")

    soup = BeautifulSoup(res.content, "html.parser")
    datas = soup.select("h4.card-text")
    for data in datas:
        print(data.get_text().strip().split(":")[1])
```
