```python
#방법2 특정한 course를 찾아오는것.
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/") 
soup = BeautifulSoup(res.content, "html.parser")

section = soup.find("ul", id="hobby_course_list")
titles = section.find_all("li", "course")

for title in titles :
    print (title.get_text().split("-")[1])
```

```python
#종현이 형님이 알려주심.. split 매서드함수 붙이기로 해결
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/") 
soup = BeautifulSoup(res.content, "html.parser")

section = soup.find_all("li", class_="course")
for sections in section:
    a = sections.get_text()
    b = a.split("-")[1]
    print(b)
```

```python
#해설
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/") 
soup = BeautifulSoup(res.content, "html.parser")

section = soup.find_all("li", class_="course")
for sections in section:
    print(sections.get_text().split("-")[-1].replace("\n"," ").replace(" ","").strip())
    
```

```python
#댓글 없애기
#해설
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/") 
soup = BeautifulSoup(res.content, "html.parser")

section = soup.find_all("li", class_="course")
for sections in section:
    a = sections.get_text().split("-")[-1].replace("\n"," ").replace(" ","").split("[")[0].strip()
    print(a)
```

```python
#댓글 없애기
#해설
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app/") 
soup = BeautifulSoup(res.content, "html.parser")

section = soup.find_all("li", class_="course")
for index, sections in enumerate(section):
    print(index+1, sections.get_text().split("-")[-1].replace("\n"," ").replace(" ","").split("[")[0].strip())
    
```
