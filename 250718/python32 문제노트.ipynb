```python
#선생님이랑 같이 1/5

import requests, openpyxl #이렇게 하는 방법도 가능! But 너무 많으면 가독성이 떨어짐..
from bs4 import BeautifulSoup

excel_file = openpyxl.Workbook() #엑셀파일을 인스턴트 객체로 만들어주는 과정!
excel_sheet = excel_file.active
excel_sheet.title = "게시글"

header = ["No","게시글제목", "댓글 수"] #행에 들어가는 값은 리스트 형태로 정리
excel_sheet.append(header)

#각각의 열의 너비를 정의
excel_sheet.column_dimensions["A"].width = 20 #디멘션 : 기준이 되는것
excel_sheet.column_dimensions["B"].width = 80 
excel_sheet.column_dimensions["C"].width = 80 #댓글들이 같이 들어오기 때문에 넓어야함.

url = "https://davelee-fun.github.io/trial/board/news.html"
index = 0
res = requests.get(url)
soup = BeautifulSoup(res.content,"html.parser")
datas = soup.select("div.list_content>div.list_item")


for data in datas :
    title = data.select_one("span.subject_fixed")
    reply_count = data.select_one("span.rSymph05") 
    link = data.select_one("a.list_subject")
    if title != None :
        index+=1
        #엑셀시트 넣는 곳!
        #print(index, title.get_text().strip(), [int(reply_count.get_text().strip())])
        data1 =[index, title.get_text().strip(), "["+reply_count.get_text().strip()+"]"] #reply_count 숫자로 가져오는 방법 -> []를 문자로 인식하게
        excel_sheet.append(data1)
        
        res_title = requests.get("https://davelee-fun.github.io/trial/board/"+link["href"])
        soup_title = BeautifulSoup(res_title.content,"html.parser")
        replies = soup_title.select("div.comment_content>div.comment_view")
        for replie in replies:
            #엑셀시트 넣는 곳 2!!
            data2 = ["","", replie.get_text().strip().replace(" ","").replace("\n","").replace("\t","")]
            #print(" ↳", replie.get_text().strip().replace(" ","").replace("\n","").replace("\t",""))#select_one("span.subject_fixed"))
            excel_sheet.append(data2)
#엑셀 파일 저장!
excel_file.save("mysite03.xlsx")
excel_file.close()

print("👏mysite03.xlsx 저장완료!!")
```

```python
#선생님이랑 같이 5/5 링크에 폰트걸기!
import requests, openpyxl 
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font, colors #블루값을 가져오기위한 colors 


wrap_alignment = Alignment(wrap_text=True) #정의해놓은 셀의 범위 초과하면 자동 줄바꿈
center_alignment = Alignment(horizontal="center", vertical="center") #가운데정렬!1
bold_font = Font(bold=True)
header_font = Font(bold=True, size=14)
normal_font = Font(bold=False) 
#폰트정의 (링크 폰트걸기_)
hyperlink_font = Font(color=colors.BLUE, underline="single") #블루폰트에 밑줄하나

excel_file = openpyxl.Workbook() 
excel_sheet = excel_file.active
excel_sheet.title = "게시글"

header = ["No","게시글제목", "댓글 수"]
excel_sheet.append(header)


for cell in excel_sheet[1]:
    cell.font = header_font
    #가운데 정렬!2
    cell.alignment = center_alignment

excel_sheet.column_dimensions["A"].width = 20 
excel_sheet.column_dimensions["B"].width = 80 
excel_sheet.column_dimensions["C"].width = 80 

url = "https://davelee-fun.github.io/trial/board/news.html"
res = requests.get(url)
soup = BeautifulSoup(res.content,"html.parser")
datas = soup.select("div.list_content>div.list_item")

row = 2
index = 0

for data in datas :
    title = data.select_one("span.subject_fixed")
    reply_count = data.select_one("span.rSymph05") 
    link = data.select_one("a.list_subject")
    if title != None :
        index+=1        
        link_url = "https://davelee-fun.github.io/trial/board/"+link["href"]
        data1 =[index, title.get_text().strip(), "["+reply_count.get_text().strip()+"]"] 
        excel_sheet.append(data1)
        excel_sheet.cell(row = row, column = 2).hyperlink = link_url
        excel_sheet.cell(row = row, column = 2).font = hyperlink_font #
        
        for cell in excel_sheet[row] : 
            if cell.column != 2 :             # 2열이 아닌 요소만 볼드 폰트, 나머지는 하이퍼링크로 이행하도록!
                cell.alignment = wrap_alignment
                cell.font = bold_font
        row+=1
        
        res_title = requests.get("https://davelee-fun.github.io/trial/board/"+link["href"])
        soup_title = BeautifulSoup(res_title.content,"html.parser")
        replies = soup_title.select("div.comment_content>div.comment_view")
        for replie in replies:
            data2 = ["","", replie.get_text().strip().replace(" ","").replace("\n","").replace("\t","")]
            #print(" ↳", replie.get_text().strip().replace(" ","").replace("\n","").replace("\t",""))#select_one("span.subject_fixed"))
            excel_sheet.append(data2)
            for cell in excel_sheet[row] :
                cell.font = normal_font
                cell.alignment = wrap_alignment 
            row+=1
            
excel_file.save("mysite14.xlsx")
excel_file.close()

print("👏mysite14.xlsx 저장완료!!")
```
