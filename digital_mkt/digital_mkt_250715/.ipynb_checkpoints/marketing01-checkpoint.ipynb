```python
# so, seleniumìœ¼ë¡œ í¬ë¡¤ë§í•˜ëŠ” ë°©ë²•!!
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
from openpyxl import Workbook


#1. selenium ê¸°ë³¸ ì˜µì…˜ ì„¤ì •
options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--headless")

driver = webdriver.Chrome(options = options)

#2. í˜ì´ì§€ ì ‘ì†
url = "https://www.kurly.com/goods/5090278"
driver.get(url)

time.sleep(5) #í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°ì‹œê°„

#3.HTML ê°€ì ¸ì˜¤ê¸°
soup = BeautifulSoup(driver.page_source, "html.parser")

# 4. ë¦¬ë·° í…ìŠ¤íŠ¸ ì¶”ì¶œí•˜ê¸°
reviews = soup.select("p.css-y49dcn.e36z05c13")
# ë¦¬ë·°ì¦ˆì˜ ê°’ì„ ì°¾ì•„ì™€ì„œ ê°’ì„ í•˜ë‚˜ì”© ì €ì¥í•˜ê¸°
review_lists =[review.get_text().strip() for review in reviews] 

driver.quit() #ì¢…ë£Œë¥¼ ì‹œí‚¤ì§€ ì•Šìœ¼ë©´ ê³„ì†í•´ì„œ í–‰ë™ì„ í•˜ê¸°ë•Œë¬¸ì— ë´‡ì€ ì´ìƒí•¨ì„ ê°ì§€í•¨.

wb = Workbook()
ws = wb.active
ws.title = "Kurly Reviews"

ws.append(["ë¦¬ë·° ë‚´ìš©"])
for item in review_lists:
    ws.append([item])

wb.save("kurly_reviews.xlsx")
print("ì—‘ì…€ì €ì¥ì™„ë£Œ : kurly_reviews.xlsx")
```
