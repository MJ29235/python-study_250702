```python
#ì´ì œ í”„ë¦¬íŒ¨ìŠ¤ë¡œ ëš«ê³ ê°€ê¸°!!!!!!
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By # 1, ë­˜ë¡œ ì°¾ì•„ì˜¬ê±´ì§€ ì„ íƒí•˜ëŠ”ê±°. (css, html ë“±)
from webdriver_manager.chrome import ChromeDriverManager
import pickle 
import time

options = Options()
options.add_argument("--start-maximized") # 3, í˜„ì¬ ìš°ë¦¬ ë¸Œë¼ìš°ì €ì— ìµœì í™”ëœ ì‚¬ì´ì¦ˆë¡œ í¬ë¡¤ë§í•˜ê²Œë”í•˜ëŠ” ì˜µì…˜
options.add_argument("window-size=1920x1080") 
options.add_argument("lang=ko_KR")

service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service, options=options)

driver.get("https://www.instagram.com/")

with open("instagram_cookies.pkl","rb") as f : # 2,ì°¾ì•„ì˜¨ ì´ë¦„ê°’ì„ ì½ì–´ì™€ì•¼í•¨ rb
    cookies = pickle.load(f) # 3, fì˜ ì‹¤í–‰ê°’ì„ ì½ì–´ì™€ë¼. cookiesëŠ” ì—¬ëŸ¬ê°œì˜ ì§ë ¬ê°’ìœ¼ë¡œ ë“¤ì–´ì™”ê¸° ë•Œë¬¸ì— ë¦¬ìŠ¤íŠ¸
for cookie in cookies:
    driver.add_cookie(cookie) # 4, ì¸ì¦ë°›ì€ ì‚¬ëŒì¸ê²ƒì²˜ëŸ¼í•˜ëŠ” í¬ë¡¤ëŸ¬
# 5, ì¿ í‚¤ê°’ì„ ê°€ì§„ ë“œë¼ì´ë²„ê°€ ìƒˆë¡œê³ ì¹¨í•œ ìƒíƒœë¡œ ì ‘ì†! -> ë¡œê·¸ì¸ì´ ì™„ë£Œëœ ìƒíƒœë¡œ ë“¤ì–´ê°
driver.refresh() #6, ë°”ë¡œ ë“¤ì–´ê°€ë©´ ì‚¬ëŒì´ ì•„ë‹Œê²ƒê°™ê¸° ë•Œë¬¸ì— í…€ì„ ë‘ !
time.sleep(5)

# 7, ì–´ë””ì„œ ì—ëŸ¬ê°€ ë°œìƒí–ˆëŠ”ê°€
try :
    not_now = driver.find_element(By.XPATH,"//button[text()='ë‚˜ì¤‘ì— í•˜ê¸°']") # 8, íŒì—…ì´ ëœ¨ë©´ ë²„íŠ¼ì˜ ì—­í• ì„ í•˜ëŠ” ìš”ì†Œì— 'ë‚˜ì¤‘ì—í•˜ê¸°' ê°€ ìˆìœ¼ë©´ ëˆŒëŸ¬ë¼
    not_now.click()
    time.sleep(2)
except :
    pass # 9, ìœ„ íŒì—…ì´ ëœ¨ì§€ ì•Šìœ¼ë©´ ê·¸ëƒ¥ ì§€ë‚˜ê°€ë¼!

keyword = "ë¬´ì‹ ì‚¬" # 10, ì›í•˜ëŠ” í‚¤ì›Œë“œë¥¼ ë„£ìœ¼ë©´ ë¨!
driver.get(f"https://www.instagram.com/explore/tags/{keyword}") 
time.sleep(5)

# 11, íŠ¹ì • ì¹´í…Œê³ ë¦¬ ì¤‘ ì–´ë–¤ ì»¨í…ì¸ ëŠ” ë„ë°°ë¥¼ í•˜ëŠ”ê²½ìš° ìˆìŒ ë„ë°° ì•ˆëœ¨ê²Œ set
post_links = set() # ì¤‘ë³µê°’ì„ ë°›ì„ ìˆ˜ ì—†ê²Œ ë§Œë“¦!
# ë¡œë”©ìŠ¤í”¼ë„ˆê°€ ëŒê³ ë‚˜ì„œ ìŠ¤í¬ë¡¤ ë‚´ë¦¬ê¸°3ë²ˆì •ë„ ë°˜ë³µí• ê²ƒ!
for _ in range(1,4) :
    links = driver.find_elements(By.TAG_NAME, "a") # 12, 3ë²ˆ ëŒë™ì•ˆ ì•µì»¤íƒœê·¸ ë‹¤ ì°¾ì•„ì˜´
    for link in links :
        href = link.get_attribute("href") # í•˜ì´í¼ë˜í¼ëŸ°ìŠ¤ì— ë§¤ì¹­ë˜ì–´ì§€ëŠ” ê°’ì„ ê°€ì ¸ì˜´
        if href and "/p/" in href : # 14, í•˜ì´í¼ë ˆí¼ëŸ°ìŠ¤ê°€ ìˆê³ , ì¶•ì•½ë˜ì–´ì§„ ë¬¸ìì—ëŠ” /p/ê°€ ë‹¤ ìˆìŒ!
            post_links.add(href)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # 15, ë†’ì´ ë‹¤ì‹œ ì¸¡ì •, ; => ì¢…ë£Œ ì‹œí€€ìŠ¤
    time.sleep(3)

results = []
#ë¦¬ìŠ¤íŠ¸ í˜•íƒœë¡œ í™•ì‹¤í•˜ê²Œ ë°”ê¾¸ê¸° -> ê°ê°ì˜ ì¸ë„¤ì¼ë¡œ ë“¤ì–´ê°„ ì‚¬ì´íŠ¸ ê°’ì„ ê°€ì ¸ì˜¤ê¸° ìœ„í•´
for url in list(post_links) [:10] : # 16, 10ê°œì”©ë§Œ ì°¾ì•„ì™€ì•¼ì•ˆë“¤í‚´
    driver.get(url)
    time.sleep(3)
    #ê°ê°ì˜ ì»¨í…ì¸ ì— ìˆëŠ” ë¬¸ì•ˆì„ ì°¾ì•„ì˜¤ê¸°
    caption = "" # 17, ìº¡ì…˜ì€ ë¹ˆ ë¬¸ìì—´ë¡œ!
    
    try :
        img_tag = driver.find_element(By.XPATH, "//img[@alt]") # 18, ê° ìš”ì†Œë“¤ì˜ ì´ë¯¸ì§€ ì°¾ì•„ì˜¤ê¸°
        caption = img_tag.get_attribute("alt").strip() # 19, altì†ì„±ì„ ì°¾ì•„ì™€ì„œ ì—¬ë°±ì„ ì—†ì• ì¤˜
        results.append({"url":url, "caption": caption})# ìº¡ì…˜ê°’ì„ ì°¾ì•„ì™”ë‹¤ë©´
    except Exception as e : ###########
        print(f"ìº¡ì…˜ ì¶”ì¶œ ì‹¤íŒ¨ : {url} | ì—ëŸ¬ : {e}") # 20, ëª‡ë²ˆì§¸ ì•µì»¤ì—ì„œ ì‹¤íŒ¨í–ˆëŠ”ì§€ ì•Œì•„ì•¼í•¨
        continue # 21, ë‹¤ìŒ ë‹¨ê³„ë¡œ ì´ë™!
        # result ë”•ì…”ë„ˆë¦¬ ê°’ ì¶”ì¶œí•˜ê¸°
driver.quit()

for r in results :
    print(r["url"])
    print(r["caption"])
    print("="*60) # êµ¬ë¶„ì„  ë§Œë“¤ì–´ì£¼ê¸°
    
```
