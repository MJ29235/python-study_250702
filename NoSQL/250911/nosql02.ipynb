```python
import requests
from bs4 import BeautifulSoup

#10개 페이지 넘기면서 크롤링!
for page_num in range(10):
    if page_num == 0 : #반복을 처음시작할때, (파라미터값이 없음!)
        res = requests.get("https://davelee-fun.github.io/")
    else : # 나머지 페이지 파라미터값 반복해서 넣어줌
        res = requests.get(f"https://davelee-fun.github.io/page{str(page_num+1)}") # 문자열의 형태로 바꿔서 안젼하게!
        
    soup = BeautifulSoup(res.content,"html.parser") # 파싱을 통해 사람이 알 수 있는 값으로 출력해줌
    #print(soup)
    data = soup.select("div.card-body") #card -body안에 있는 모든 데이터를 가져옴.
    #print(data)

    #mongo db 에 저장할때 나눠서 저장하기 위해 data의 값들을 나눠줘야함 -> for in 문
    for item in data:
        category = item.select_one("h2.card-title").get_text().replace("관련 상품 추천","").strip() 
        # h2태그의 card-title을 찾아옴 # 코드는 빼고 타이틀만 찾아오기
        product = item.select_one("h4.card-text").get_text().replace("관련 상품 추천","").strip() 
        print(category,product) 
```

```python
############ MongoDB로 값을 넣고 전환해보기! #############
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient # pymongo 라이브러리 가져오기!

client = MongoClient("mongodb://localhost:27017") # 접속
db = client["test"] # DB만들기
collection = db.daveshop #컬렉션 생성 

for page_num in range(10):
    if page_num == 0:
        res = requests.get("https://davelee-fun.github.io/")
    else:
        res = requests.get(f"https://davelee-fun.github.io/page{str(page_num+1)}")
        
    soup = BeautifulSoup(res.content, "html.parser")
    data = soup.select("div.card-body")

    for item in data:
        category = item.select_one("h2.card-title").get_text().replace("관련 상품 추천", "").strip()
        product = item.select_one("h4.card-text").get_text().replace("관련 상품 추천", "").strip()
        #print(category, product)
        #Mongo DB 에 넣을때는 키와 밸류값을 가진 딕셔너리 형태로 넣어줘야함! dict()
        data = dict()
        data["category"] = category #열 만들어주기
        data["product"] = product
        collection.insert_one(data)

#데이터 누수 방지
client.close()
```

```python
#예제 카테고리 필드에 있는 모든 카테고리 종류를 조회해주세요.
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
db = client["test"]
collection = db.daveshop
```

```python
categories = collection.distinct("category")
categories
```

```python
# 모든 상품 (product) 조회
products = collection.find({},{"_id":0,"product":1})
#list(products)
for product in products:
    print(product["product"])
```

```python
#매트리스 커버 카테고리의 모든 상품을 조회해주세요
Pipeline = [
    {"$match":{"category":"*/매트리스커버*/"}}
]

products = collection.aggregate(Pipeline)
for product in products:
    print(product)
```

```python
products=collection.find({"categort":"매트리스커버"},{"_id":0,"product":1})

for product in products:
    print(product["product"])
```

```python
pipeline = [
    {"$match":{"category":"매트리스커버"}},
    {"$project":{"_id":0,"product":1}}
]

for product in collection.aggregate(pipeline):
    print(product["product"])
```

```python
# 각 카테고리별 상품수를 조회해주세요
pipeline = [
    {"$group": {"_id": "$category", "count": {"$sum": 1}}},
    {"$project": {"_id": 0, "category": "$_id", "count": 1}}
]

for product in collection.aggregate(pipeline):
    print(product["category"], product["count"])
```

```python
pipeline=[
    {"$group":{"_id":"$category","count":{"$sum":1}}}
]

for category_count in collection.aggregate(pipeline):
    print(category_count["_id"], category_count["count"])
```

```python
#상품명에 핑크가 포함된 상품명을 조회
products = collection.find({"product":{"$regex":".*핑크.*"}},{"_id":0,"product":1})
for product in products:
    print(product)
```
