```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")

db = client.sample_mflix
movies = db.movies

print(movies)
```

```python
#aggregate 문법 python에서 사용해보기
```

```python
#find() => aggregation Framework 문법을 사용하면 pipeline 생성해서 복잡한 데이터 처리를 순차적으로 세팅할 수 있음

# 1. $match => 특정 조건에 해당되는 문서를 찾아오고자 할 떄

for movie in movies.aggregate(
    [{"$match" : {"genres":"Action"}},{"$limit":1} ]
):
    print(movie)
```

```python
# 2. $group => SQL 에서의 Group BY와 동일한 기능
for movie in movies.aggregate(
    [{"$group" : {"_id":"$directors","count":{"$sum":1}}}]
):
    print(movie)
```

```python
# ==
pipeline = [{"$group" : {"_id":"$directors","count":{"$sum":1}}}]
for movie in movies.aggregate(pipeline):
    print(movie)
```

```python
# $sort는 문서의 정렬을 하고싶을 때 사용. 복수의 정렬을 할 수 있음. 다만, '좌측'부터 정렬 방식이 적용됨! 
pipeline = [
    {"$sort":{"title":1}},{"$limit":3}
]

for movie in movies.aggregate(pipeline):
    print(movie)

```

```python
# project : 특정 원하는 값을 출력하고자 할때 사용
pipeline = [
    {"$project":{"_id":0,"title":1,"genres":1}},{"$limit":3}
]

for movie in movies.aggregate(pipeline):
    print(movie)
    
```

```python
# $unwind : 배열 안에 입력된 값을 추출해서 출력하고자 할
pipeline = [
    {"$unwind":"$genres"},{"$limit":3}
]

for movie in movies.aggregate(pipeline):
    print(movie)
    
```

```python
# $group 과 집계연산자 응용 패턴
pipeline=[
    {"$group":{"_id":"$directors","count":{"$sum":1},"average_rating":{"$avg":"$imdb.rating"}}}
]
for movie in movies.aggregate(pipeline) :
    print(movie)
```

```python
# 컬렉션에 있는 영화의 갯수
count = movies.count_documents({}) # 아무것도 없는 {}는 조건이 존재하지 않는 모든 값을 찾아옴
print("영화수 : ",count)
```

```python
# 평균 영화 길이를 찾으세요
pipeline=[
    {"$group":{"_id":0,"average_length":{"$avg":"$runtime"}}}
]
for movie in movies.aggregate(pipeline) :
    print("평균 영화 길이 :",movie)
result = list(movies.aggregate([
    {"$group":{"_id":0,"average_length":{"$avg":"$runtime"}}}
]))
result[0]["average_length"]
```

```python
# 각 장르에 대한 영화 갯수를 계산하세요
# pipeline=[
#     {"$group":{"_id":"$genres","count":{"$sum":1}}}
# ]
# for movie in movies.aggregate(pipeline) :
#     print("장르별 영화 갯수 :",movie)

################액션 
genre_counts=list(
movies.aggregate([
    {"$unwind":"$genres"},
    {"$group":{"_id":"$genres","count":{"$sum":1}}}
]))
for genre in genre_counts:
    print(f"장르 : {genre["_id"]},영화수 {genre["count"]}개")
```

```python
# 2015년 이후에 개봉한 영화를 제목으로 정렬하여 나열해주세요.
# for movie in movies.aggregate(
#     [{"$match" : {year:{"$gt":2015}}},{"$sort":{"title":1}} ]
# ):
#     print(movie)

post_2015_movies = movies.find({"year":{"$gt":2015}}).sort("title")
for movie in post_2015_movies:
    print(f"영화제목 : {movie['title']}")

#aggregation Framework 버전!
pipeline = [
    {"$match":{"year":{"$gt":2015}}}, {"$sort":{"title":1}}, {"$project":{"_id":0,"title":1}}
]
for movie in movies.aggregate(pipeline):
    print(f"영화제목 : {movie['title']}")
```

```python
# 가장 많은 영화를 제작하는 상위 5개국을 찾아주세요
# for movie in movies.aggregate(
#     {"$group" : {"_id":"countries","count":{"$sum":1}}}, {"$sort":{"$countries":1}}
# ):
#     print(movie)

####################
pipeline = [
    {"$unwind":"$countries"},
    {"$group":{"_id":"$countries","count":{"$sum":1}}},
    {"$sort":{"count":-1}},
    {"$limit":5}
]
top_countries = movies.aggregate(pipeline)
for country in top_countries :
    print(f"국가 : {country['_id']}, 영화수 :{country['count']}개")
```

```python
#2000년 이후 영화의 연도별 평균 IMDB 평점을 찾으세요 

pipeline = [
    {"$match":{"year":{"$gte":2000}}},
    {"$group":{"_id":"$year","avgrating":{"$avg":"$imdb.rating"}}},{"$sort": {"_id": 1}}, {"$project":{"_id":0,"year":"$_id","avg_rating": {"$round": ["$avgrating", 2]}}}
]
for movie in movies.aggregate(pipeline):
    print(movie)
```
