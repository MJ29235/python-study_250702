```python
import pandas as pd

doc = pd.read_csv(
    r"C:\Users\pmj96\PythonBasic\pandas\00_PANDASPLOTLY_FUNCODING_FIRSTDATA_V3_1\00_PANDASPLOTLY_FUNCODING_FIRSTDATA\COVID-19-master\csse_covid_19_data\csse_covid_19_daily_reports\04-01-2020.csv",
    encoding="utf-8-sig"
)
```

```python
doc2 = doc.corr(numeric_only=True)
```

```python
doc2
```

```python
# 

import plotly.express as px

fig_px=px.imshow(
    doc2,
    #text_auto=True,
    text_auto=".2f",
    aspect ="auto",
    title="Heatmap"
)

fig_px.show()
```

```python
# graph_object

import plotly.graph_objects as go

fig = go.Figure(
    
)

fig.add_trace(
    go.Heatmap(
        x=doc2.columns,
        y=doc2.index,
        z=doc2.values,
        #colorscale="Viridis",
        colorscale="Cividis",
        colorbar={
            "title": "value"
        },
        text=doc2.values,
        texttemplate="%{z:.2f}",
        hovertemplate="X:%{x}<br>Y:%{y}<br>Value:%{z:.2f}<br>Custom Info"
    )
)

fig.update_layout(
    title="Heatmap"
)

fig.show()
```

```python
import plotly.express as px

fig_px = px.scatter(
    doc,
    x="Confirmed",
    y="Active",
    title="Scatter Plot"
)
fig_px.update_layout(
    template="plotly_dark"
)

fig_px.show()
```
