```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait as WW   # ✅ 추가
from selenium.webdriver.support import expected_conditions as EC # ✅ 추가
from webdriver_manager.chrome import ChromeDriverManager
import pickle
import time
import json, re

def _expand_more(driver, timeout=2):
    xps = [
        "//button//*[contains(.,'더보기')]",
        "//button[normalize-space()='더 보기']",
        "//button//*[translate(., 'MORE', 'more')='more']",
    ]
    for xp in xps:
        try:
            btn = WW(driver, timeout).until(EC.element_to_be_clickable((By.XPATH, xp)))
            btn.click()
            time.sleep(0.2)
            return True
        except Exception:
            continue
    return False

def _clean(s):
    return re.sub(r"\n{3,}", "\n\n", s.strip()) if s else ""

def caption_from_dom(driver):
    sels = [
        (By.CSS_SELECTOR, "h1._ap3a"),
        (By.CSS_SELECTOR, "h1[dir='auto']"),
        (By.CSS_SELECTOR, "span[style*='line-height']"),
    ]
    _expand_more(driver)
    for by_, sel in sels:
        try:
            e = WW(driver, 2).until(EC.presence_of_element_located((by_, sel)))
            t = _clean(e.text)
            if t:
                return t, "dom"
        except Exception:
            continue
    return "", ""

def caption_from_jsonld(driver):
    try:
        for sc in driver.find_elements(By.CSS_SELECTOR, "script[type='application/ld+json']"):
            data = json.loads(sc.get_attribute("textContent") or "{}")
            for k in ("caption", "articleBody", "description"):
                if isinstance(data, dict) and k in data:
                    t = _clean(data[k])
                    if t:
                        return t, "jsonld"
    except Exception:
        pass
    return "", ""

def caption_from_meta(driver):
    for s in [
        "meta[property='og:description']",
        "meta[name='description']",
    ]:
        try:
            t = driver.find_element(By.CSS_SELECTOR, s).get_attribute("content")
            if t:
                return _clean(t), "meta"
        except Exception:
            continue
    return "", ""

def caption_from_alt(driver):
    try:
        img = driver.find_element(By.CSS_SELECTOR, "img[alt]")
        return _clean(img.get_attribute("alt")), "alt"
    except Exception:
        return "", ""

def get_caption_anyway(driver):
    for func in [caption_from_dom, caption_from_jsonld, caption_from_meta, caption_from_alt]:
        t, src = func(driver)
        if t:
            return t, src
    return "", ""

options = Options()
options.add_argument("--start-maximized")
options.add_argument("window-size=1920x1080")
options.add_argument("lang=ko_KR")
# options.add_argument("--headless=new")  # 필요 시 사용

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://www.instagram.com")

# --- 쿠키 로드 ---
with open("instagram_cookies.pkl", "rb") as f:
    cookies = pickle.load(f)
for cookie in cookies:
    try:
        driver.add_cookie(cookie)
    except Exception:
        pass

driver.refresh()
time.sleep(3)

# --- 팝업 닫기 ---
try:
    not_now = driver.find_element(By.XPATH, "//button[text()='나중에 하기']")
    not_now.click()
    time.sleep(1)
except Exception:
    pass

# --- 해시태그 페이지 이동 ---
keyword = "도시락"
driver.get(f"https://www.instagram.com/explore/tags/{keyword}")
time.sleep(3)

# --- 게시물 링크 수집 ---
post_links = set()
for _ in range(3):
    links = driver.find_elements(By.TAG_NAME, "a")
    for link in links:
        href = link.get_attribute("href")
        # ✅ 리일스까지 포함하려면 아래 조건 사용 (원래대로면 '/p/'만 유지)
        if href and ("/p/" in href or "/reel/" in href):
            post_links.add(href)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# --- 결과 수집 ---
results = []
for url in list(post_links)[:5]:
    driver.get(url)

    # 콘텐츠 로딩 대기(없으면 그냥 진행)
    try:
        caption, source = get_caption_anyway(driver)
        results.append({"url": url, "caption": caption, "source": source})
        WW(driver, 6).until(EC.presence_of_element_located((By.CSS_SELECTOR, "article")))
    except Exception:
        pass
    time.sleep(1)

    # 기본값
    user_id = ""
    caption_text = ""
    likes_text = ""
    date_iso = ""

    try:
        # 1) 사용자 ID
        try:
            user_id = driver.find_element(
                By.CSS_SELECTOR, "span._ap3a._aaco._aacw._aacx._aad7._aade"
            ).text.strip()
        except Exception:
            pass

        # 2) 본문 캡션 (우선 h1._ap3a → 없으면 line-height span → 없으면 img[alt])
        try:
            caption_text = driver.find_element(
                By.CSS_SELECTOR, "h1._ap3a"
            ).text.strip()
        except Exception:
            try:
                caption_text = driver.find_element(
                    By.CSS_SELECTOR, "span[style*='line-height']"
                ).text.strip()
            except Exception:
                pass
        if not caption_text:
            try:
                alt_img = driver.find_element(By.CSS_SELECTOR, "img[alt]")
                caption_text = (alt_img.get_attribute("alt") or "").strip()
            except Exception:
                pass

        # 3) 좋아요 수 (예: "좋아요 2790개")
        try:
            likes_text = driver.find_element(
                By.CSS_SELECTOR,
                "span.x193iq5w.xeuugli.x1fj9vlw.x13faqbe.x1vvkbs.xt0psk2.x1i0vuye.xvs91rp.x1s688f.x5n08af.x10wh9bi.xpm28yp.x8viiok.x1o7cslx"
            ).text.strip()
        except Exception:
            pass

        # 4) 날짜 (ISO datetime 속성)
        try:
            date_el = driver.find_element(By.CSS_SELECTOR, "time.x1p4m5qa")
            date_iso = (date_el.get_attribute("datetime") or "").strip()
        except Exception:
            pass

        results.append({
            "url": url,
            "user_id": user_id,
            "caption": caption_text,  # 기존 caption 키 유지
            "likes": likes_text,
            "date": date_iso,
        })

    except Exception as e:
        print(f"캡션 추출 실패 : {url} | 에러 : {e}")
        # 실패해도 URL만 저장
        results.append({
            "url": url,
            "user_id": user_id,
            "caption": caption_text,
            "likes": likes_text,
            "date": date_iso,
        })
        continue

driver.quit()

# --- 출력 ---
for r in results:
    print(r["url"])
    print("ID:", r["user_id"] or "(없음)")
    print("본문:", r["caption"] or "(없음)")
    print("좋아요:", r["likes"] or "(없음)")
    print("날짜:", r["date"] or "(없음)")
    print("=" * 60)
```
